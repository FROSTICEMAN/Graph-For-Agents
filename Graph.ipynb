```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes for agents and key steps
G.add_node("User")
G.add_node("Code Creation Agent")
G.add_node("Generated Code")
G.add_node("Testing Agent")
G.add_node("Test Passed?")
G.add_node("Provide Feedback to User")
G.add_node("Feedback to Code Creation Agent")

# Add edges to represent the workflow
G.add_edges_from([
    ("User", "Code Creation Agent"),
    ("Code Creation Agent", "Generated Code"),
    ("Generated Code", "Testing Agent"),
    ("Testing Agent", "Test Passed?"),
    ("Test Passed?", "Provide Feedback to User"),
    ("Test Passed?", "Feedback to Code Creation Agent"),
    ("Feedback to Code Creation Agent", "Code Creation Agent")
])

# Draw the graph
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)  # Layout for positioning nodes
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", arrows=True)

# Display the graph
plt.title("Agent Workflow: Code Creation and Testing")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes for agents and key steps
G.add_node("User")
G.add_node("Code Creation Agent")
G.add_node("Generated Code")
G.add_node("Testing Agent")
G.add_node("Test Passed?")
G.add_node("Provide Feedback to User")
G.add_node("Feedback to Code Creation Agent")

# Add edges to represent the workflow
G.add_edges_from([
    ("User", "Code Creation Agent"),
    ("Code Creation Agent", "Generated Code"),
    ("Generated Code", "Testing Agent"),
    ("Testing Agent", "Test Passed?"),
    ("Test Passed?", "Provide Feedback to User"),
    ("Test Passed?", "Feedback to Code Creation Agent"),
    ("Feedback to Code Creation Agent", "Code Creation Agent")
])

# Define node positions in a straight line
pos = {
    "User": (0, 0),
    "Code Creation Agent": (1, 0),
    "Generated Code": (2, 0),
    "Testing Agent": (3, 0),
    "Test Passed?": (4, 0),
    "Provide Feedback to User": (5, 0),
    "Feedback to Code Creation Agent": (5, -1),
}

# Draw edges and labels first
plt.figure(figsize=(12, 4))
nx.draw_networkx_edges(G, pos, arrows=True)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold")

# Get node positions as coordinates
node_coords = [pos[node] for node in G.nodes()]

# Draw square nodes using scatter (squared marker 's')
plt.scatter(
    [x for x, y in node_coords], 
    [y for x, y in node_coords], 
    s=3000, color="skyblue", marker="s"
)

# Display the graph
plt.title("Agent Workflow: Code Creation and Testing (Square Nodes)")
plt.show()
```

```python
import schemdraw
from schemdraw.flow import *

with schemdraw.Drawing() as d:
    d+= Start().label("Start")
    d+= Arrow().down(d.unit/2)
    
    #Input the string 
    d+= Data(w = 4).label("Enter a string:\n string")
    d+= Arrow().down(d.unit/2)
    
    #Reverse the string
    d+= Box(w = 4).label("Reverse the string:\n reverse_string")
    d+= Arrow().down(d.unit/2)
    
    #Check if string and reverse_string are same
    d+= (decision := Decision(w = 5, h= 5,
                       S = "True",
                        E = "False").label("Is \n string\n == \nreverse_string?"))
    
    #If True
    d+= Arrow().length(d.unit/2)
    d+= (true := Box(w = 5).label("string is a palindrome."))
    d+= Arrow().length(d.unit/2)
    
    #End program
    d+= (end := Ellipse().label("End"))
    
    #If False. Start the arrow from East of decision box
    d+= Arrow().right(d.unit).at(decision.E)
    
    #false is referring to the box where string is not a palindrome.
    d+= (false := Box(w = 5).label("string is not\n a palindrome."))
    
    #Add a downward arrow from the South of false box 
    d+= Arrow().down(d.unit*2.5).at(false.S)
    
    #Extend the arrow to reach the end of the program
    d+= Arrow().left(d.unit*2.15)
    d.save("output/palindrome flowchart.jpeg", dpi = 300)
```
